```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

# Data points
x = [1, 2, 3, 5, 6, 7, 8, 9, 10, 12, 13, 14, 15, 16, 18, 19, 21, 22]
y = [100, 90, 80, 60, 60, 55, 60, 65, 70, 70, 75, 76, 78, 79, 90, 99, 99, 100]

# Fit a polynomial model of degree 3
mymodel = np.poly1d(np.polyfit(x, y, 3))

# Generate a line for plotting
myline = np.linspace(1, 22, 100)

# Create the plot
plt.scatter(x, y, color='blue', label='Data points')
plt.plot(myline, mymodel(myline), color='red', label='Polynomial fit')
plt.title('Polynomial Fit of Degree 3')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()

# Calculate and print R² score
r2 = r2_score(y, mymodel(x))
print(f'R² Score: {r2:.4f}')
```
